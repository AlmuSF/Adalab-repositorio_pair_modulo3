```python
# Crear una función para agregar la función que estamos agregando - como hacer un diagnóstico - detectar que podemos meter en la función
```

```python
def data_inspect(df,head,tail,sample):
    print('The number of rows and columns is:')
    print(df.shape)
    print('---------------------------------------------------------------------')
    print('These are the columns that we have')
    print(df.columns)
    print('---------------------------------------------------------------------')
    print('The variables are of these types:')
    print(df.dtypes)
    print('---------------------------------------------------------------------')
    print('Quantity of missings in each column/variable:')
    print(df.isnull().sum())
    print('---------------------------------------------------------------------')
    print('Number of duplicated rows:')
    print(df.duplicated().sum())
    print('---------------------------------------------------------------------')
    print(f'Description of groups included in each type of data')
    for column in df.columns:
        print(df.describe(include = df[column].dtype).T)
    print('---------------------------------------------------------------------')
    print(f'First {head} rows')
    print(df.head(head))
    print('---------------------------------------------------------------------')
    print(f'Last {tail} rows')
    print(df.tail(tail))
    print('---------------------------------------------------------------------')
    print(f'A random sample of {sample}')
    print(df.sample(n=sample))
    
```

```python
# df = name of the data frame
df = df # add the name of the DataFrame
head = 10 # add the number of the first rows that you want to see.
tail = 10 # add the number of the first rows that you want to see.
random = 5 # add the number of random rows that you want to see.

data_inspect(df,head, tail, random)
```
