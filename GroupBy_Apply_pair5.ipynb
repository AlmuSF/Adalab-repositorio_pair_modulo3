```python
# 1. Agrupamos por pais de origen 'origin_country' y sumamos las 'units_sold'.

ventas_paises= df_merchant.groupby('origin_country')['units_sold'].sum().reset_index().sort_values(by='units_sold', ascending=False)
ventas_paises
```

```python
# Estadisticas de precio por pais de envio. Precio promedio y precio maximo.
```

```python
df_precio_pais= df_merchant.groupby('origin_country')['price'].agg(['mean','max'])
df_precio_pais
```

```python
df_precio_pais= df_precio_pais.sort_values(by='max')
df_precio_pais
```

Obersvaciones:
- CN tiene un precio max aislado muy alto en comparacion con su media. Su media se ubica en el medio entre los paises de medias mas altas y bajas. 

- SG se desmarca del resto de medias destacando por ser un pais mucho más barato y teniendo su precio max muy cerca de su propia media.

```python
#5. Reemplazo de valores.
```

```python
diccionario_mapa = {0: "No", 1: "Si"}
columnas_reemplazar= ['uses_ad_boosts','badge_local_product','badge_product_quality','badge_fast_shipping','shipping_is_express']

for col in columnas_reemplazar:

    df_merchant[col] = df_merchant[col].map(diccionario_mapa)
    df_merchant[col].unique()
```

```python
#6. clasificación por calidad del producto. Sobreescribir columnas.
```

```python
df_merchant["badge_product_quality"] = df_merchant["badge_product_quality"].replace('No', 'Buena')
df_merchant['badge_product_quality'].unique()

df_merchant.head()
```
